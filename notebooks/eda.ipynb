### Analisis Exploratorio de Datos

##### Importacion de librerias a usar

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Cargado del Dataset

```python
# Url del dataset
dir = "dataset_alquiler.csv"
df = pd.read_csv(dir)
df.head() # no utilizamos el print por que asi se ve mas estetico la presentacion
```

#### Descripcion y Resumen Estadistico

```python
print("##Informacion General##")
df.info()
```

```python
print("##Estadisticas Dataset##")
df.describe()
```

- Podemos Notar que en las columnas hora, dia_semana y total_alquileres existen datos faltantes , asi que trataremos de impugnarlos o rellenarlos 

#### Columna-Hora

```python
filas_faltantes_hora = df[df['hora'].isnull()]
print("\nFilas con valores faltantes en la columna 'hora':")
filas_faltantes_hora
```

```python
# Viendo la falta de datos en los segmentos continuos
print('#Hora 313-319#')
print(df['hora'][313:320])
print("----------")
print('#Hora 598:604')
print(df['hora'][597:606]) 
# Notamos que los datos siguen una forma de serie 
```

```python
# Rellenar los valores faltantes en el primer segmento (índices 313 a 318)
df.loc[313:318, 'hora'] = range(13, 13 + 6)

# Rellenar los valores faltantes en el segundo segmento (índices 599 a 603)
df.loc[599:603, 'hora'] = range(2, 2 + 5)

# Verificar los cambios
print("\nValores de 'hora' después de la imputación en el primer segmento:")
print(df['hora'][310:321])

print("\nValores de 'hora' después de la imputación en el segundo segmento:")
print(df['hora'][595:607])

# Volver a verificar si quedan valores nulos en 'hora'
print("\nCantidad de valores nulos restantes en la columna 'hora':")
print(df['hora'].isnull().sum())
```

#### Columna Dia_Semana

```python
filas_faltantes_dia_semana = df[df['dia_semana'].isnull()]
print("\nFilas con valores faltantes en la columna 'dia_semana':")
filas_faltantes_dia_semana
```

```python
# Viendo la falta de datos en los segmentos continuos
print('#dia_semana 3-9#')
print(df['dia_semana'][3:10])
print("----------")
print('#dia_semana 128-134')
print(df['dia_semana'][127:135]) # Los valores son determinables y podemos ir viendo que en los 
# intervalos que no tiene datos se pueden repetir los valores
```

```python
# Rellenar los valores faltantes en el rango de índices 3 a 9 con el valor 6.0
df.loc[3:9, 'dia_semana'] = 6.0
# Rellenar los valores faltantes en el rango de índices 128 a 133 con el valor 4.0
df.loc[128:133, 'dia_semana'] = 4.0
print("\nCantidad de valores nulos restantes en la columna 'dia_semana':")
print(df['dia_semana'].isnull().sum())
```

#### Columna Total Alquileres

```python
filas_faltantes_total_alquileres = df[df['total_alquileres'].isnull()]
print("\nFilas con valores faltantes en la columna 'total_alquileres':")
filas_faltantes_total_alquileres
```

```python
print("---#Rango de Valores alquileres 85-99#---")
print(df['total_alquileres'][85:100])
print("---#Rango de Valores alquileres 5490-5500#---")
print(df['total_alquileres'][5490:5501]) # Los valores son numericos y algo dificiles de inferir , 
#para no perde los demas datos lo remplazaremos con el promedio de los datos
```

```python
# Calcular la media de la columna 'total_alquileres'
media_total_alquileres = df['total_alquileres'].mean()
# Imputar los valores faltantes con la media
df['total_alquileres'] = df['total_alquileres'].fillna(media_total_alquileres)
# Verificar la cantidad de valores nulos restantes en 'total_alquileres'
print(f"\nCantidad de valores nulos restantes en la columna 'total_alquileres': {df['total_alquileres'].isnull().sum()}")
```

#### Analisis Variable Objetivo

```python
# Distribución de la Variable Objetivo
plt.figure(figsize=(12, 5))
sns.histplot(df['total_alquileres'], kde=True)
plt.title('Distribución de Total de Alquileres')
plt.xlabel('Número Total de Alquileres')
plt.ylabel('Frecuencia')
plt.show()

#  Diagrama de Caja de la Variable Objetivo
plt.figure(figsize=(8, 5))
sns.boxplot(y=df['total_alquileres'])
plt.title('Diagrama de Caja de Total de Alquileres')
plt.ylabel('Número Total de Alquileres')
plt.show()

# Estadísticas Resumen de la Variable Objetivo
print("\n### Estadísticas Resumen de la Variable Objetivo:")
print(df['total_alquileres'].describe())
```

#### Podemos observar que tiene una gran cantidad de outliers , por el momento no los imputaremos pero lo tendremos en consideracion

# Análisis de las Variables Numéricas

```python
print("Analisis de las Variables Numéricas")

numerical_features = ['temperatura', 'sensacion_termica', 'humedad', 'velocidad_viento']

for col in numerical_features:
    print(f"\n### Analisis de la variable numérica: {col}")

    # Distribucion de la Variable Numerica
    plt.figure(figsize=(12, 5))
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribucion de {col}')
    plt.xlabel(col)
    plt.ylabel('Frecuencia')
    plt.show()

    # Relacion con la Variable Objetivo
    plt.figure(figsize=(12, 5))
    sns.scatterplot(x=df[col], y=df['total_alquileres'])
    plt.title(f'Relacion entre {col} y Total de Alquileres')
    plt.xlabel(col)
    plt.ylabel('Total de Alquileres')
    plt.show()

# Matriz de Correlacion de Variables Numericas y la Variable Objetivo
correlation_matrix = df[numerical_features + ['total_alquileres']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correlacion de Variables Numéricas y Total de Alquileresf')
plt.show()
```
